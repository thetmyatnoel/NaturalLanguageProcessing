# 7-1 Sigmoid function


```python
import matplotlib.pyplot as plt 
import numpy as np
def sigmoid(x, alpha, beta):
    return 1.0 / (1.0 + np.exp(-(alpha + beta*x)))
z = np.arange(-7, 7, 0.1)
c0 = [sigmoid(x, alpha=0, beta=1) for x in z] 
plt.plot(z, c0, label='alpha=0, beta=1', linestyle='--')
c1 = [sigmoid(x, alpha=1, beta=1) for x in z] 
plt.plot(z, c1, label='alpha=1, beta=1')
c2 = [sigmoid(x, alpha=1, beta=2) for x in z] 
plt.plot(z, c2, label='alpha=1, beta=2')
c3 = [sigmoid(x, alpha=0, beta=2) for x in z] 
plt.plot(z, c3, label='alpha=0, beta=2')
plt.axvline(0.0, color='k') 
plt.ylim(-0.1, 1.1) 
plt.xlabel('z') 
plt.ylabel('$\phi (z)$')
# y축의 눈금과 격자선 
plt.yticks([0.0, 0.5, 1.0]) 
ax = plt.gca() 
ax.yaxis.grid(True)
plt.legend(loc='best') 
plt.tight_layout()
plt.show() 
```

# 7-2 Sigmoid function & Sigmoid Derivative

```python
import numpy as np
import matplotlib.pyplot as plt
def sigmoid(x, diff=False): 
  if diff:
   return sigmoid(x)*(1-sigmoid(x)) 
  else:
   return 1.0 / (1.0 + np.exp(-x)) 
x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x, False) 
y2 = sigmoid(x, True)
plt.figure(figsize=(8, 6))
plt.ylim(-0.2, 1.0)
plt.title("Sigmoid & Derivative", fontsize=30) 
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15, rotation=0)
plt.axvline(0.0, color='gray', linestyle="--", alpha=0.8) 
plt.axhline(0.0, color='gray', linestyle="--", alpha=0.8) 
plt.plot(x, y1, c="blue", label="sigmoid")
plt.plot(x, y2, c="purple", label="derivative") 
plt.legend(loc="upper left")
plt.yticks([0.0, 0.25, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.show()
```

# 7-3 tanh function & tanh Derivative

```python
import numpy as np
import matplotlib.pyplot as plt
def tanh(x) :
  return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
# 탄젠트 도함수
def tanh_derivative(x):
  tan = tanh(x) 
  return 1-tan**2
x_range = np.arange(-10, 10, 0.2)
y_range = np.array([tanh(x) for x in x_range])
y_prime = np.array([tanh_derivative(x) for x in x_range])
plt.plot(x_range, y_range, label='tanh') 
plt.plot(x_range, y_prime, label='tanh_prime') 
plt.grid(which='major')
plt.legend()
plt.show()
```

# 7-4 ELU & Swish function

```python
import numpy as np
import matplotlib.pyplot as plt
def ELU(x, a=1):
  return (x>0)*x + (x<=0)*(a*(np.exp(x) - 1))
def sigmoid(a, x):
  return 1.0 / (1.0 + np.exp(-a*x))
def swish_function(a, x): return x * sigmoid(a, x)
x = np.arange(-5.0, 5.0, 0.1) 
a=1
y1 = ELU(x, a)
y2 = swish_function(a, x)
plt.figure(figsize=(8, 6)) 
plt.ylim(-2, 7)
plt.title("ELU & Swish", fontsize=30)
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15, rotation=0)
plt.axvline(0.0, color='gray', linestyle="--", alpha=0.8) 
plt.axhline(0.0, color='gray', linestyle="--", alpha=0.8) 
plt.plot(x, y1, c="blue", linestyle="--", label="ELU") 
plt.plot(x, y2, c="green", label="Swish") 
plt.legend(loc="upper left")
plt.show()
```
